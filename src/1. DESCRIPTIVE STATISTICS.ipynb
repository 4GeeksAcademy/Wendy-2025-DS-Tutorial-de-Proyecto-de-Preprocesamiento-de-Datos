### 12. BEST PRACTICES AND RECOMMENDATIONS FOR DESCRIPTIVE DATA ANALYSIS

- Convert categorical columns to numeric using label encoding, if is necesary.
- Remove outliers without losing relevant data:
This part can change many times, because we have to do it by looking at our graphs at the end. We can only remove outliers if the values are origininally numerical. Never if the values were originally categorical.
- Irrelevant columns should be removed, because they may cause noise, confusions or be processed in error.
- Some columns could be indexed, as it is not relevant. But may be useful for future references.
- We can remove no necesary columns when we have other/s that may give us the same information.
- We have to remove not relevant information, to have a agile processing.
- Already calculated values have to be removed, because they're detrimental to training.
- Variables with a single unique value, null value, duplicated value, or value equal to 0, should be removed.
- Remove outliers, better if we use the interquartile range method (IQR).
